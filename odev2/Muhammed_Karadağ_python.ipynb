```python
#Gün değerlendirme

#hata verildiği zaman tekrardan başlamak için döngüye katıyoruz.
while True:
  #kullanıcıdan verileri alıyoruz
  uyku = input("Bugün kaç saat uyudunuz ? -> ")
  ders = input("Bugün kaç saat ders yaptınız = -> ")
  serbest = input("Bugün kaç saat serbest zaman geçirdiniz ? -> ")

  #yazdığı verilerin sayısal değer olup olmadığına bakıyoruz
  if not uyku.isdigit() or not ders.isdigit() or not serbest.isdigit():
    print("Lütfen sayısal veri giriniz...")
    continue

  #kullanıcıdan aldığımızın değerlerin sayısal veri olduğunu öğrenince
  #verileri tam sayı olarak ayarlıyoruz
  uyku = int(uyku)
  ders = int(ders)
  serbest = int(serbest)

  #kullanıcının verdiği verilerin pozitif sayı olup olmadığına bakıyoruz
  if serbest < 0 or ders < 0 or uyku < 0:
    print("Lütfen pozitif sayı giriniz...")
    continue

  #kullanıcıdan alınan verilerin tam bir gün olup olmadığına bakıyoruz
  if serbest + ders + uyku != 24:
    print("Lütfen doğru zaman biçimi giriniz...")
    continue

  #kullanıcının uykusunun değerine bakıyoruz
  if uyku < 6 : uyku_deger = "yetersiz uyku"
  elif uyku < 8: uyku_deger = "yeterli uyku"
  else: uyku_deger = "aşırı uyku"

  #kullanıcının ders çalışma verilerine bakıyoruz
  if ders < 6: ders_deger = "normal çalışma"
  else: ders_deger = "aşırı çalışma"

  #kullanıcının uyku ve çalışma değerlerini kullanıcıya sunuyoruz
  print(f"uyku değeriniz = {uyku_deger}")
  print(f"çalışma değeriniz = {ders_deger}")

  #kullanıcının değerlerine bakıp verimli gün geçirip geçirmediğine bakıyoruz
  if uyku_deger == "yetersiz uyku" and ders_deger == "aşırı çalışma":
    print("verimsiz bir gün geçiriyorsunuz")
  else:
    print("verimli bir gün geçiriyorsunuz")
  break
```
